```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset

df = pd.read_csv('Data/KWO (1st time Test).csv')


# print(df.head())

# make all the same length
df = df[:600]

df.dropna(inplace=True)

df['hours'], df['minutes'], df['seconds'] = map(df['Time'].str.split(':').str.get, [0, 1, 2])

df['total_time'] = df['hours'].astype(int) * 3600 + df['minutes'].astype(int) * 60 + df['seconds'].astype(int)

initial_time = df['total_time'][0]

df['total_time'] = df['total_time'] - initial_time

plt.plot(df['total_time'], df['Resistance '], color='red', label='Resistance')
plt.title('Resistance vs Time --> KWO (1st time Test)')

plt.legend(loc='upper left')

#create intersection at the vertical line




plt.show()




```

```python
plt.plot(df['total_time'], df['relative humidity [%RH]'], color='blue', label='relative humidity [%RH]')
plt.legend(loc='upper left')
plt.title('relative humidity [%RH] vs Time')
```

```python
#second csv 

df_2 = pd.read_csv('Data/KWO Acetone High Humidity (2nd time test).csv')

df_2 = df_2[:600]

df_2.dropna(inplace=True)

df_2['hours'], df_2['minutes'], df_2['seconds'] = map(df_2['Time'].str.split(':').str.get, [0, 1, 2])

# Check if 'Resistance' column exists in the DataFrame

df_2['total_time'] = df_2['hours'].astype(int) * 3600 + df_2['minutes'].astype(int) * 60 + df_2['seconds'].astype(int)

initial_time = df_2['total_time'][0]

df_2['total_time'] = df_2['total_time'] - initial_time

plt.plot(df_2['total_time'], df_2['Resistance '], color='red', label='Resistance')

plt.legend(loc='upper left')

plt.title('Resistance vs Time --> KWO Acetone High Humidity (2nd time test)')

plt.show()


```

```python
#second humidity plot

plt.plot(df_2['total_time'], df_2['relative humidity [%RH]'], color='blue', label='relative humidity [%RH]')

plt.legend(loc='upper left')

plt.title('relative humidity [%RH] vs Time --> KWO Acetone High Humidity (2nd time test)')

plt.show()
```

```python
#third csv

from sklearn.linear_model import LinearRegression

df_3 = pd.read_csv('Data/KWO Acetone - Low Humidity (3rd time Check).csv')

df_3 = df_3[:600]

df_3['Resistance '] = df_3['Resistance ']/1000000

df_3.dropna(inplace=True)

df_3['hours'], df_3['minutes'], df_3['seconds'] = map(df_3['Time'].str.split(':').str.get, [0, 1, 2])

# Check if 'Resistance' column exists in the DataFrame

df_3['total_time'] = df_3['hours'].astype(int) * 3600 + df_3['minutes'].astype(int) * 60 + df_3['seconds'].astype(int)

initial_time = df_3['total_time'][0]

df_3['total_time'] = df_3['total_time'] - initial_time

plt.plot(df_3['total_time'], df_3['Resistance '], color='red', label='Resistance')

#linear regression

x = df_3['total_time'].values.reshape(-1, 1)

y = df_3['Resistance '].values.reshape(-1, 1)

reg = LinearRegression()

reg.fit(x, y)

predictions = reg.predict(x)

plt.plot(x, predictions, color='blue', linewidth=1, label='Linear Regression')\

plt.ylabel('Resistance (Mega Ohms)')

#print the coefficients m and b

print('m = ', reg.coef_)

print('b = ', reg.intercept_)







plt.legend(loc='upper left')

plt.title('Resistance vs Time --> KWO Acetone - Low Humidity (3rd time Check)')

plt.show()
```

```python
#third humidity plot

plt.plot(df_3['total_time'], df_3['relative humidity [%RH]'], color='blue', label='relative humidity [%RH]')

plt.legend(loc='upper left')

plt.title('relative humidity [%RH] vs Time --> KWO Acetone - Low Humidity (3rd time Check)')

#linear regression

x = df_3['total_time'].values.reshape(-1, 1)

y = df_3['relative humidity [%RH]'].values.reshape(-1, 1)

reg = LinearRegression()

reg.fit(x, y)

predictions = reg.predict(x)

plt.plot(x, predictions, color='red', linewidth=1, label='Linear Regression')

plt.ylabel('relative humidity [%RH]')

#print the coefficients m and b

print('m = ', reg.coef_)

print('b = ', reg.intercept_)





plt.legend(loc='upper left')



plt.show()
```
